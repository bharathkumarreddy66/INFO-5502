```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("lab-assignment-04-student-performance.csv")
df
```

```python
df.describe()
```

```python
df.columns
```

```python
ax=sns.histplot(df['sh'])
plt.xlabel('study hours')
ax.bar_label(ax.containers[0])
plt.title("Count plot STUDY HOURS")
```

```python
sns.set_style('darkgrid')
df['counts']=df.groupby(['sh','esp'])['esp'].transform('count')
ax=sns.factorplot(x='sh',y='counts',hue='esp',data=df,hue_order=['Best','Vg','Good','Pass'],kind='bar')
plt.title('Students percentage vs study hours')
plt.xlabel('study hours')
plt.ylabel('counts')
```

```python
dum=df.copy()
for c,v in dum.iteritems():
  if len(dum[c].unique())==1:#ckecking for uniquiness of coulumn
    to_drop=c
dum.drop(to_drop,inplace=True,axis=1)
corr = dum.apply(lambda x: x.factorize()[0]).corr()
plt.figure(figsize=(15,15))
sns.heatmap(corr, annot=True, cmap="Reds")
plt.title(' Heatmap')
```
