```python
c = x['OverAllRating'].shape[0]
e = y['OverAllRating'].shape[0]
alpha = 0.05
mc = np.mean(x['OverAllRating'])
me = np.mean(y['OverAllRating'])

vc = np.var(x['OverAllRating'])
ve = np.var(y['OverAllRating'])


pv = np.sqrt(vc/mc + ve/me)

T = (mc-me)/np.sqrt(vc/mc + ve/me)
p_value = norm.sf(T)*2
Z  = norm.ppf(1-alpha/2)
p = Z*pv
CI = [(mc - me) - p, (mc - me) + p]


print("Test Statistics: ", T)
print("Z-critical: ", Z)
print("P_value: ", p_value)
print("Confidence Interval: ", np.round(CI,2))
```
